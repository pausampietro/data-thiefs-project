```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from lxml import html
import urllib.request
import re
```

```python
#url = ["https://www.barcelona-airport.com/eng/departures.php?tp=0", 
 #      "https://www.barcelona-airport.com/eng/departures.php?tp=6",
  #     "https://www.barcelona-airport.com/eng/departures.php?tp=12", 
  #     "https://www.barcelona-airport.com/eng/departures.php?tp=18"]
url = "https://www.barcelona-airport.com/eng/departures.php?tp=6"    
departure = requests.get(url).content
soup = BeautifulSoup(departure, "lxml")
soup
```

```python
details = [element.text for element in soup.find_all(id="fdest")]
```

```python
details1 = [element.text for element in soup.find_all(id="fair")]
details2 = [element.text for element in soup.find_all(id="fhour")]
details3 = [element.text for element in soup.find_all(id="fnum")]
details_strip = [i.strip() for i in details]
details_split = [re.split('\(', i) for i in details_strip]
#details_split_destination = [i.strip().split(" ")[0] for i in details]
#details_split_destination
```

```python
details_split
```

```python
departures = df = pd.DataFrame({'destination' : details_split,
                                'airline' : details1,
                                'flight_num' : details3,
                                'hour' : details2 }, 
                                columns=['destination', 'airline', "flight_num", 'hour'])
```

```python
departures[['destination','airport_code']] = pd.DataFrame(departures.destination.values.tolist(), index= departures.index)
```

```python
departures['airport_code'].replace(regex=True,inplace=True,to_replace=r'\)',value=r'')
```
