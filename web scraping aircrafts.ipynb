```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

url = 'https://en.wikipedia.org/wiki/Fuel_economy_in_aircraft'
html = requests.get(url).content
html[0:600]
```

```python
soup = BeautifulSoup(html, "lxml")
soup
```

```python
table = soup.find_all('table',{'class':'wikitable sortable'})[1]
rows = table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
```

```python
colnames = ['Model', 'First flight', 'Seats', 'Sector', 'nan1', 'Fuel burn', 'nan2', 'Fuel efficiency per seat']
data = rows[1:]

regional_flights = pd.DataFrame(data, columns=colnames)
regional_flights
```
