```python
import pandas as pd
import sqlalchemy
import pymysql
from sqlalchemy import create_engine
```

```python
aircrafts = pd.read_csv('https://raw.githubusercontent.com/jpatokal/openflights/master/data/planes.dat',
            header=None, sep=',', engine='python')
```

```python
aircrafts = aircrafts.rename(columns= {0: 'aircraft_name',1:'iata',2:'icao'})
```

# Manual corrections of data

```python
aircrafts['icao'].iloc[12]='A330'
aircrafts['icao'].iloc[15]='A340'
aircrafts['icao'].iloc[20]='A350'
aircrafts['icao'].iloc[23]='A380'
```

```python
aircrafts[aircrafts['icao']=='\\N']
```

```python
aircrafts = aircrafts.drop(labels = aircrafts[aircrafts['icao']=='\\N'].index, axis=0)
```

```python
aircrafts
```

```python
aircrafts = aircrafts.replace('\\N','ND')
aircrafts = aircrafts.reset_index(drop=True)
aircrafts
```
